```python
import pandas as pd
import numpy as n
```

cleaning the data - convert fields names and values into lowercase and replace spaces with underscore sign

```python
df = pd.read_csv('https://raw.githubusercontent.com/alexeygrigorev/datasets/master/laptops.csv')
df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
df.head()
```

```python
TextFields = list(df.dtypes[df.dtypes == 'object'].index)
for s in TextFields:
    df[s] = df[s].str.lower().str.replace(' ','_')
df.head()
```

Exploratory data analysis

```python
Total_datasets=len(df)
Testing_Dataset_count=int(0.2*Total_datasets)
Validation_Dataset_Count=int(0.2*Total_datasets)
Train_Dataset_Count = Total_datasets - Testing_Dataset_count-Validation_Dataset_Count
Total_datasets,Testing_Dataset_count,Validation_Dataset_Count,Train_Dataset_Count
```
